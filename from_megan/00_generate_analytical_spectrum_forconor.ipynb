# ANALYTICAL APPROACH TO DYNAMIC TOPOGRAPHY

```python
########################################
########### IMPORT LIBRARIES ###########
########################################

import os.path
import sys
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python

########################################
########### PLOT PARAMETERS ############
########################################

plt.rcParams['figure.figsize'] = [15, 10]
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)
plt.rcParams['font.size'] = 20

########################################
########### CONSTANTS ##################
########################################

Radius=6.371e6  # m, radius of the earth (6371 km)
r=2.890e6       # m, depth (y-axis)      (3000 km)
c=1.500e6       # m, parabola minimum    (1500 km) # c in manuscript

rho_a=3330     # kg m-3
rho_w=1000     # kg m-3
rho_delta=2330 # kg m-3

# This will sweep through multiple density parabolas
#b_start=0        # kg m-5, shape of parabola # b in manuscript
#b_stop=4.250e-12   # kg m-5, shape of parabola # b in manuscript
#b_inc=0.250e-12     # kg m-5, shape of parabola # b in manuscript

# This will just use the density plotted in the paper (blue line in Figure 18d)
b_start=1.000e-12        # kg m-5, shape of parabola # b in manuscript
b_stop=1.250e-12   # kg m-5, shape of parabola # b in manuscript
b_inc=0.250e-12     # kg m-5, shape of parabola # b in manuscript

r_start=0        # m, depth of parabola (   0 km)
r_stop=3000000   # m, depth of parabola (3000 km)
r_inc=100000      # m, depth of parabola (  10 km)

###########################################################################

# define arrays

# -- arrays for density anomaly
array_x=[]
array_r=[]

# -- arrays for analytical spectrum

array_l=[]
array_k=[]
array_DT_pow=[]
array_DT_amp=[]
array_exp=[]

  
```

```python
# GENERATE SURFACE TOPOGRAPHY EXPONENTIALS 

for l in range(1,51,1):

    outfile3 = open("outputfolder/exp_l_%.0f.txt" %(l), "w")

    for r in range(r_start,r_stop,r_inc):                    # r = radius of the mantle

        # 1. define terms
        lamb=(2.*np.pi*Radius)/(np.sqrt(l*(l+1.))) # wavelength
        k=2.*np.pi/lamb                            # wavenumber

        # 2. calculate the exponential
        exponential=np.exp(-k*r)

        # 3. save to file
        outfile3.write(str(r) + " " + str(exponential) + "\n")    

    outfile3.close()   


for b in np.arange(b_start,b_stop,b_inc): 
    # GENERATE DENSITY ANOMALY -- this is calculated in kg/m3, to replicate the diagram
    outfile2 = open("outputfolder/Density_b_%.2f.txt" %(b*1e12), "w")
    for r in range(r_start,r_stop,r_inc): # depth to the base of the mantle, m

        # 1. approximate density anomaly using a parabola
        x=b*(r-c)**2.  

        # 2. save to array
        array_x.append(x)
        array_r.append(r)

        # 3. save to file
        outfile2.write(str(x) + " " + str(r) + "\n")
    outfile2.close()       

    # GENERATE SPECTRUM

    outfile1 = open("outputfolder/DT_b_%.2f.txt" %(b*1e12), "w")
    for l in range(1,51,1):

        lamb=(2.*np.pi*Radius)/(np.sqrt(l*(l+1.))) # wavelength
        k=2.*np.pi/lamb # wavenumber

        VarA=(1/(rho_delta)) # (1/rho_a-rho_w)
        VarB=((b*c**2.)/k)-((2.*b*c)/k**2.)+((2.*b)/k**3.)-np.exp(-k*r)*(((b*(r-c)**2.)/k)+((2.*b*(r-c))/k**2.)+((2.*b)/k**3.))

        # calculate dynamic topography (power and amplitude)
        DT_pow=(VarA*VarB)**2 /1e6 # change from m to km using 1e6
        DT_amp=VarA*VarB
        
        # save to array
        array_l.append(l)
        array_k.append(k)
        array_DT_pow.append(DT_pow)
        array_DT_amp.append(DT_amp)

        outfile1.write(str(l) + " " + str(k) + " " + str(DT_pow) + " " + str(DT_amp) + "\n")

    outfile1.close()     
```

```python
#######################################
##### PLOT DENSITY ANOMALY ############
#######################################

# i.e. blue line in Figure 18d

plt.figure()
plt.rcParams['figure.figsize'] = [5, 5] # width, height
plt.scatter(array_x,array_r)
plt.xlabel('Density Anomaly, kg m-3')
plt.ylabel('Depth')
```

```python
#######################################
##### PLOT ANALYICAL SPECTRUM #########
#######################################

# i.e. pink line in Figure 18e

plt.figure()
plt.rcParams['figure.figsize'] = [5, 5] # width, height
plt.yscale("log")
plt.scatter(array_l,array_DT_pow)
plt.xlabel('Degree, l')
plt.ylabel('Power, km**2')
plt.ylim(0.0001,1000)
plt.savefig('3.png')
```
